### AHP 层次分析法

包括目标、准则和方案。

对问题进行问题分析，例如为了消除数量级的影响而进行归一化。

现实中，指标对于评价的重要性不同，例如粉丝数比稿件数更能体现up主的优秀。

指标数值乘以权重再相加，以区分重要性

> e.g. 0.4 * 粉丝数 + 0. 1* 稿件数量

但是权重是由人为主观设定的，这也是层次分析法的缺点。

换做是你，觉得下面谁才是人类呢？

![](image-2.png )

如何科学地设置权重？有个比较好的方案是，**对指标的重要性进行两两比较，构造判断矩阵，从而求出权重**

矩阵中元素 $a_{i,j}$ 的意义是，第i个指标相对第j个指标的重要程度。
