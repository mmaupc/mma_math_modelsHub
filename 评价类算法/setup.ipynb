### AHP 层次分析法

AHP，即Analytic Hierarchy Process，是一种将复杂决策问题系统化的方法。它将问题拆解为三个清晰的层次：

- 方案层 (Alternatives)： 有哪些可供选择的方案？
- 准则层 (Criteria)： 评价的标准有哪些？
- 目标层 (Goal)： 我们要解决什么问题？

合在一起就是ACG文化（确信）

使用~~ACG~~AHP的原因在于，现实中不同指标的重要性不一样。在B站看up主的影响力时，我们普遍地认为粉丝数比稿件数更能体现up主的含金量。最直观的方法就是指标数值乘以权重再相加，以区分重要性，例如：

> e.g. 0.4 * 粉丝数 + 0. 1* 稿件数量

但是权重是由人为主观设定的，有很强的主观性。

那么如何科学地设置权重？有个比较好的方案是，**对指标的重要性进行两两比较，构造判断矩阵，从而求出权重**

我们不直接决定粉丝数占多少权重，二十构造一个判断矩阵，对指标的重要性进行两两对比。

- 矩阵中元素 $a_{i,j}$ 的意义是，第i个指标相对第j个指标的重要程度。

通过这种方式，我们将复杂的总体转换为简单的二元比较，利用数学方法例如特征值法求出权重，从而使得决策更加科学和客观。
