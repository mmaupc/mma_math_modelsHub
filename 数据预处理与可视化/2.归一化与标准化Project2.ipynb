```python
import numpy as np
import pandas as pd
import os
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from scipy.stats import boxcox

# 1. 读取数据
file_path = r'D:\数模协会\学术部\Examples1.csv'
df = pd.read_csv(file_path)
print("原始数据预览：")
print(df.head())

# 创建一个新的DataFrame来保存处理后的结果
result = pd.DataFrame()

# 2. 处理 '归一化' 列（Min-Max归一化）
data_norm = df['归一化'].values.reshape(-1, 1)  # sklearn需要二维输入

# 2.1 使用封装函数
scaler = MinMaxScaler()
norm_scaled = scaler.fit_transform(data_norm).flatten()
result['归一化_原始'] = df['归一化']
result['归一化_库函数'] = norm_scaled

# 2.2 手动计算归一化
min_val = df['归一化'].min()
max_val = df['归一化'].max()
norm_manual = (df['归一化'] - min_val) / (max_val - min_val)
result['归一化_手动'] = norm_manual

# 验证手动与库函数结果是否一致
print("\n归一化一致性检查（差值绝对值最大）:", np.max(np.abs(norm_scaled - norm_manual)))

# 3. 处理 '标准化' 列（Z-score标准化）
data_std = df['标准化'].values.reshape(-1, 1)

# 3.1 使用封装函数
scaler_std = StandardScaler()
std_scaled = scaler_std.fit_transform(data_std).flatten()
result['标准化_原始'] = df['标准化']
result['标准化_库函数'] = std_scaled

# 3.2 手动计算标准化
mean_val = df['标准化'].mean()
std_val = df['标准化'].std(ddof=0)
std_manual = (df['标准化'] - mean_val) / std_val
result['标准化_手动'] = std_manual

print("标准化一致性检查（差值绝对值最大）:", np.max(np.abs(std_scaled - std_manual)))

# 4. 处理 '对数变换' 列（自然对数）
log_transformed = np.log(df['对数变换'])
result['对数变换_原始'] = df['对数变换']
result['对数变换_处理后'] = log_transformed

# 5. 处理 'Box-Cox变换' 列
# 使用scipy的boxcox函数，返回变换后的数据和最优lambda
boxcox_transformed, fitted_lambda = boxcox(df['Box-Cox变换'])
result['Box-Cox变换_原始'] = df['Box-Cox变换']
result['Box-Cox变换_处理后'] = boxcox_transformed
result['Box-Cox变换_lambda'] = fitted_lambda  # 记录拟合的lambda值

print(f"\nBox-Cox变换拟合的lambda: {fitted_lambda:.4f}")

# 6. 保存结果到新CSV
output_path = os.path.join(r'D:\数模协会\学术部', 'result2.csv')
result.to_csv(output_path, index=False, encoding='utf-8')
print(f"\n处理完成，结果已保存至：{output_path}")
```
