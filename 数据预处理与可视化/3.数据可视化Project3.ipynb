```python
# 7. 热力图
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# 设置随机种子，保证每次运行结果一致
np.random.seed(42)

# 生成模拟数据（1000个样本）
n_samples = 1000
data = {
    'CreditScore': np.random.randint(300, 850, n_samples),           # 信用分 300-850
    'Age': np.random.randint(18, 70, n_samples),                     # 年龄 18-70
    'Tenure': np.random.randint(0, 10, n_samples),                   # 在任时间 0-9年
    'Balance': np.random.uniform(0, 250000, n_samples).round(2),     # 余额 0-25万
    'NumOfProducts': np.random.randint(1, 5, n_samples),             # 产品数量 1-4
    'EstimatedSalary': np.random.uniform(10000, 200000, n_samples).round(2)  # 预估薪资 1万-20万
}
a = pd.DataFrame(data)

# 选取用于相关性分析的列
new_a = a[["CreditScore","Age","Tenure","Balance","NumOfProducts","EstimatedSalary"]]

# 计算相关系数矩阵，保留4位小数
data_corr = new_a.corr().round(4) # method='pearson'/ method='spearman'

# 绘制热力图
plt.figure(figsize=(8, 6))
sns.heatmap(data_corr,
            annot=True,          # 显示数值
            cmap='Blues',        # 色块颜色
            linewidths=1,
            linecolor='black',
            fmt='.4f'             # 显示格式，与round(4)对应
)
plt.title('Correlation Heatmap')  # 标题
plt.tight_layout()                 # 自动调整布局
plt.show()
```

```python
# 9. 组合图表
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 设置中文字体，防止乱码
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 设置随机种子，使结果可重现
np.random.seed(42)

# 生成数据
months = np.arange(1, 13)  # 1月到12月
sales = np.random.randint(50, 100, size=12) * 10   # 销售额 500~1000
costs = sales * np.random.uniform(0.5, 0.8, size=12)  # 成本为销售额的50%~80%
profit_margin = (sales - costs) / sales              # 利润率
growth_rate = np.random.uniform(-0.05, 0.1, size=12) # 月增长率

# 整理为DataFrame
df = pd.DataFrame({
    '月份': months,
    '销售额': sales,
    '成本': costs,
    '利润率': profit_margin,
    '增长率': growth_rate
})

# 创建图形和左Y轴
fig, ax1 = plt.subplots(figsize=(10, 6))

# 柱状图位置
x = np.arange(len(months))
width = 0.35

# 绘制柱状图（两个系列）
bars1 = ax1.bar(x - width/2, df['销售额'], width, label='销售额', color='skyblue')
bars2 = ax1.bar(x + width/2, df['成本'], width, label='成本', color='salmon')

# 设置左Y轴
ax1.set_xlabel('月份')
ax1.set_ylabel('金额（元）')
ax1.set_title('月度销售额、成本、利润率及增长率')
ax1.set_xticks(x)
ax1.set_xticklabels(months)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# 创建右Y轴（共享X轴）
ax2 = ax1.twinx()

# 绘制折线图（两条线）
line1 = ax2.plot(x, df['利润率'], color='green', marker='o', linewidth=2, markersize=8, label='利润率')
line2 = ax2.plot(x, df['增长率'], color='purple', marker='s', linewidth=2, markersize=8, label='增长率')

# 设置右Y轴
ax2.set_ylabel('比率')
ax2.set_ylim(-0.2, 1.0)

# 合并图例（柱状图 + 折线图）
handles = [bars1, bars2] + line1 + line2
labels = ['销售额', '成本', '利润率', '增长率']
ax1.legend(handles, labels, loc='upper left', frameon=True)

# 调整布局并显示
plt.tight_layout()
plt.show()
```

```python
# 10. 成对关系图
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# 设置随机种子，保证结果可复现
np.random.seed(42)

# 定义三种鱼类及其样本数
species_list = ['Bream', 'Perch', 'Pike']
n_samples = [50, 60, 40]  # 不同物种样本数

# 为每个物种定义特征的均值和标准差（模拟真实分布）
# 特征顺序：Weight(g), Length1(cm), Length2(cm), Length3(cm), Height(cm), Width(cm)
species_params = {
    'Bream': {
        'mean': [600, 30, 35, 40, 15, 8],
        'std':  [100, 3,  4,  5,  2,  1.5]
    },
    'Perch': {
        'mean': [400, 25, 28, 32, 10, 5],
        'std':  [80,  2.5, 3,  4,  1.5, 1]
    },
    'Pike': {
        'mean': [800, 45, 50, 55, 8, 6],
        'std':  [150, 5,  6,  7,  1,  1.2]
    }
}

# 生成数据
data = []
for species, n in zip(species_list, n_samples):
    params = species_params[species]
    # 生成特征值（正态分布，并确保为正）
    features = np.abs(np.random.normal(loc=params['mean'], scale=params['std'], size=(n, 6)))
    # 构建DataFrame块
    df_species = pd.DataFrame(features, columns=['Weight', 'Length1', 'Length2', 'Length3', 'Height', 'Width'])
    df_species['Species'] = species
    data.append(df_species)

# 合并所有物种
fish = pd.concat(data, ignore_index=True)

# 查看前几行
print(fish.head())

# 筛选三种鱼
fish1 = fish[fish['Species'].isin(['Bream', 'Perch', 'Pike'])]

# 设置主题并绘制pairplot
sns.set_style("white")
sns.pairplot(fish1, hue='Species', palette='husl', height=2)
plt.show()
```

```python
# 12. 小提琴图
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import warnings
warnings.filterwarnings("ignore")

# 恢复默认 matplotlib 设置
mpl.rcParams.update(mpl.rcParamsDefault)
sns.set_style("white")

# 设置中文字体和解决负号显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 设置随机种子，保证可复现
np.random.seed(42)

# 定义年龄组（按顺序）
age_groups = ['16-19', '20-24', '25-34', '35-44', '45-54', '55-64', '65+']
genders = ['Male', 'Female']

# 为每个年龄-性别组合生成模拟失业数据
data_list = []
for age in age_groups:
    for gender in genders:
        # 设置不同组的均值和标准差（模拟现实：年轻人失业率高，老年人次之，中年人低）
        if age in ['16-19', '20-24']:
            mean = 12.0 if gender == 'Male' else 10.5
        elif age in ['25-34', '35-44']:
            mean = 6.0 if gender == 'Male' else 5.2
        elif age in ['45-54', '55-64']:
            mean = 5.5 if gender == 'Male' else 4.8
        else:  # 65+
            mean = 7.0 if gender == 'Male' else 6.2
        std = 2.0  # 标准差

        # 生成该组的样本数（每组约 50-80 个）
        n = np.random.randint(50, 80)
        # 生成正态分布数据，并确保非负
        unemployed = np.abs(np.random.normal(mean, std, n))

        # 添加到列表
        for value in unemployed:
            data_list.append({'Age': age, 'Gender': gender, 'Unemployed': value})

# 创建 DataFrame
employment = pd.DataFrame(data_list)

# 绘制小提琴图
plt.figure(figsize=(12, 6))
sns.violinplot(
    x=employment.Age,
    y=employment.Unemployed,
    hue=employment.Gender,
    palette='Set2',
    split=False,          # 是否分开显示，默认False并排显示
    inner='box',      # 显示四分位数线
    cut=0                  # 限制小提琴范围不超出数据范围
)
plt.xlabel('年龄组', fontsize=12)
plt.ylabel('失业人数（千人）', fontsize=12)
plt.title('不同年龄组与性别的失业人数分布', fontsize=14)
plt.legend(title='性别')
plt.tight_layout()
plt.show()
```

```python
# 13. 折线图进阶
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 随机种子
np.random.seed(42)

# 生成时间序列
periods = pd.date_range('2005-01-01', '2015-12-01', freq='M')

data = []

for i, p in enumerate(periods):

    # 构造趋势：2009-2010峰值
    trend = 100000 + 500000 * np.exp(-((i-65)/20)**2)

    # 每个月生成多个样本
    for _ in range(6):   # 样本少 → CI更大

        # 增大波动 → 误差条更长
        men = trend + np.random.normal(0, 70000)
        women = trend * 0.75 + np.random.normal(0, 60000)

        data.append([p.year + p.month/12, men, 'Men'])
        data.append([p.year + p.month/12, women, 'Women'])

employment = pd.DataFrame(data, columns=['Period','Unemployed','Gender'])
# 绘图
plt.figure(figsize=(14,7))

plt.style.use('ggplot')

plt.gcf().text(.2, .84, 'GENDER', fontsize=40, color='Black')

sns.set(rc={
    'xtick.labelsize':17,
    'ytick.labelsize':10,
    'axes.labelsize':15,
    'axes.grid':False
})

sns.lineplot(
    x='Period',
    y='Unemployed',
    hue='Gender',
    style='Gender',
    data=employment,
    dashes=False,
    palette='CMRmap',
    err_style='bars',
    ci=99,
    markers=['o', '>']
)

plt.xlabel("Period")
plt.ylabel("Unemployed")

plt.show()
```

```python
# 14. 联合分布图
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 设置随机种子，保证结果可复现
np.random.seed(42)

# 生成模拟数据
n_samples = 1000
bmi = np.random.normal(26, 5, n_samples)          # BMI 均值26，标准差5
bmi = np.clip(bmi, 10, 50)                         # 限制在合理范围 10~50

# 费用与 BMI 呈线性关系：费用 = 2000 * (BMI - 10) + 基础费用 + 噪声
# 设置基础费用和斜率使数据有合理范围
base_charge = 2000
charges = base_charge + 300 * (bmi - 10) + np.random.normal(0, 1000, n_samples)
charges = np.abs(charges)  # 确保费用为正

# 创建 DataFrame
insurance = pd.DataFrame({'bmi': bmi, 'charges': charges})

# 绘制联合分布图
my_dpi = 96
plt.figure(figsize=(480/my_dpi, 480/my_dpi), dpi=my_dpi)
sns.jointplot(x=insurance["bmi"], y=insurance["charges"], kind='hex')
plt.show()
```

```python
# 15. 直方分布图
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import warnings
warnings.filterwarnings("ignore")

# 恢复默认 matplotlib 设置
mpl.rcParams.update(mpl.rcParamsDefault)

# 设置随机种子，保证结果可复现
np.random.seed(42)

# 生成模拟数据
n_samples = 2000

housing = pd.DataFrame({
    # 经度：加州范围约 -124 到 -114
    'longitude': np.random.uniform(-124, -114, n_samples),
    # 纬度：加州范围约 32 到 42
    'latitude': np.random.uniform(32, 42, n_samples),
    # 房屋年龄中位数：1~52 年（均匀分布）
    'housing_median_age': np.random.randint(1, 53, n_samples),
    # 总房间数：对数正态分布，模拟右偏
    'total_rooms': np.random.lognormal(mean=6, sigma=1, size=n_samples).astype(int),
    # 人口：对数正态分布
    'population': np.random.lognormal(mean=5, sigma=1.5, size=n_samples).astype(int),
    # 家庭数：对数正态分布
    'households': np.random.lognormal(mean=4, sigma=1, size=n_samples).astype(int),
    # 收入中位数：对数正态分布，范围约 0.5~15
    'median_income': np.random.lognormal(mean=1.5, sigma=0.5, size=n_samples),
    # 房价中位数：对数正态分布，单位美元（乘1000使数值合理）
    'median_house_value': np.random.lognormal(mean=12, sigma=0.8, size=n_samples) * 1000
})

# 创建子图
fig1, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 20))

# 绘制分布图
sns.histplot(housing["longitude"], bins=20, kde=True, ax=axes[0, 0])
sns.histplot(housing["latitude"], bins=20, kde=True, ax=axes[0, 1])
sns.histplot(housing["housing_median_age"], bins=20, kde=True, ax=axes[0, 2])
sns.histplot(housing["total_rooms"], bins=20, kde=True, ax=axes[1, 0])
sns.histplot(housing["population"], bins=20, kde=True, ax=axes[1, 1])
sns.histplot(housing["households"], bins=20, kde=True, ax=axes[1, 2])
sns.histplot(housing["median_income"], bins=20, kde=True, ax=axes[2, 0])
sns.histplot(housing["median_house_value"], bins=20, kde=True, ax=axes[2, 1])
sns.histplot(housing["median_house_value"], bins=20, kde=True, ax=axes[2, 2])  # 重复绘制，与原代码一致

plt.tight_layout()
plt.show()
```

```python
# 16. 词云图
# 需要事先安装缺失的库
# !pip install jieba wordcloud matplotlib
# !pip install wordcloud pillow matplotlib

import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np

# 文本分词
text = """
“花无凋零之日，意无传递之时，爱情亘古不变，紫罗兰与世长存。”
“我想知道「我爱你」的含义。”
“你可曾窥视过我的眼瞳。 若未有过，便请来窥视一下如何。 我的眼瞳虽被夜色遮蔽，夜空之中却闪耀着繁星。 还望前来默然窥视。我的眼瞳中浮现之物、映照之物，倘若你认为其美丽。那便是你爱着我的证据。”
“我想，我们正是在伤害与被伤害的循环往复中，渐渐变得温柔的”
“传达心意可真是难啊， 明明只是想说谢谢而已”
“正因为深爱着。赌上了一切、赌上了人生，抹杀了自己，想着要去守护。”
“爱…就是，想着要…保护生命中最重要的人。”
“薇尔莉特…你要活下去自由地活下去我真心…爱着你”
“擅长为了他人而扼杀自己，虽然不知道至今为止杀掉了多少，恐怕双手已经被自己的血染遍了。”
"""

# 将“爱”作为一个独立词条加入词典
jieba.add_word('爱', freq=1000000)

# 分词及停用词过滤
words = jieba.lcut(text)
stopwords = set([
    '我', '你', '着', '了', '的', '是', '在', '和', '与', '也', '就', '都', '而', '及',
    '“', '”', '…', '，', '。', '、', ' ', '\n', '\t', '这', '那', '它', '它们', '我们',
    '你们', '他们', '自己', '被', '把', '让', '给', '所', '等', '并', '或', '不', '无',
    '要', '之', '中', '可', '时'
])
filtered_words = [word for word in words if word not in stopwords and word.strip() != '']
word_str = ' '.join(filtered_words)

# 生成心形蒙版
size = 800
x = np.linspace(-1.8, 1.8, size)
y = np.linspace(-1.5, 1.8, size)
xx, yy = np.meshgrid(x, y)

# 心形方程：(x^2+y^2-1)^3 - x^2*y^3 <= 0 表示内部（尖点朝上）
heart_eq = (xx**2 + yy**2 - 1)**3 - xx**2 * yy**3
mask = np.where(heart_eq <= 0, 0, 255).astype(np.uint8)

# 垂直翻转蒙版，使心形尖点朝下
mask = np.flipud(mask)

# 指定中文字体路径
font_path = 'simkai.ttf'

# 生成词云
wc = WordCloud(
    font_path=font_path,
    width=size,
    height=size,
    background_color='white',
    mask=mask,
    max_words=100,
    collocations=False,
    stopwords=stopwords
).generate(word_str)

"""
# 若不用蒙版，则将相关代码删除，直接生成词云
wc = WordCloud(
    font_path=font_path,
    width=800,
    height=600,
    background_color='white',
    max_words=100,
    collocations=False,
    stopwords=stopwords
).generate(word_str)
"""

# 显示
plt.figure(figsize=(10, 10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 17. 词云图进阶
# 需要事先安装缺失的库
# !pip install wordcloud pillow matplotlib

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image

# ------------------ 定义文本内容 ------------------
strengths_text = """
Inclusiveness and Adaptability: Based on Client Performance Evaluation Model (CPEM), we have meticulously selected the most suitable client for our project. Through assessing the status quo of wildlife protection and execution capabilities with a total of 12 sub-indicators, we ensure a tailored approach that lays a strong foundation for the project’s success.
Strong Client-Project Alignment: The project’s rationale is clearly articulated, demonstrating the Chinese government’s readiness in terms of power, resources, and interest. This strategic alignment is essential for the project’s effectiveness and long-term sustainability.
Comprehensive Project Design: The project design emphasizes digital innovation, integrating blockchain, machine learning, big data, and virtual reality technology. It also considers the approach to monitoring and managing the whole industrial chain of illegal wildlife trade, providing a multifaceted approach to tackling the illicit activity.
Data-Driven Decision Making: The project is further supported by data-driven methodologies, including CPEM and IPM. Empirical evidence from news reports and literature underpins the project’s design and strategies, ensuring decisions are informed by robust data.
"""

weaknesses_text = """
Data Accuracy Dependence: The project’s effectiveness relies heavily on the availability and accuracy of statistical data. The complexity of the model and the challenge of data collection mean that the outcomes are highly dependent on the quality of the data available.
Complex Computational Process: The project’s computational processes are complex and resource-intensive, requiring significant effort and computational resources to execute. This complexity could be a barrier to rapid implementation and scalability.
Lack of Consideration for Cyclical Changes: The current model does not fully account for the cyclical nature of illegal wildlife trade, which could impact the project’s effectiveness over time. Seasonal variations in demand and enforcement efforts are not captured, potentially affecting the model’s long-term predictive accuracy.
"""

# ------------------ 自定义停用词 ------------------
custom_stopwords = set(STOPWORDS) | {
    'based', 'using', 'also', 'including', 'project', 'data', 'model',
    'approach', 'efforts', 'changes', 'trade', 'wildlife', 'illegal',
    'provide', 'providing', 'ensure', 'ensuring', 'demonstrate', 'through',
    'with', 'for', 'that', 'this', 'these', 'those', 'could', 'may', 'will',
    'must', 'need', 'new', 'one', 'two', 'three', 'four', 'first', 'second',
    'third', 'fourth', 'fifth', 'time', 'long', 'term', 'over', 'into',
    'within', 'without', 'includes', 'well', 'et', 'al', 'etc', 'e', 'g',
    'i', 'e', 'cpem', 'ipm', 'client', 'design', 'decision', 'driven', 
    's', "'s", 'accuracy', 'effectiveness', 'computational', 'resource',
}

# ------------------ 加载掩码图片（请修改为你的实际路径）------------------
mask_up = np.array(Image.open(r"D:\数模协会\学术部\thumb_up.png").convert("L"))
mask_down = np.array(Image.open(r"D:\数模协会\学术部\thumb_down.png").convert("L"))

# ------------------ 自定义颜色函数（固定浅色）------------------
def light_green_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return (144, 238, 144)  # LightGreen 的 RGB 值

def light_red_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return (255, 182, 193)  # LightCoral 的 RGB 值

# ------------------ 生成优势词云（浅绿色文字）------------------
wc_strengths = WordCloud(
    background_color='white',
    mask=mask_up,
    stopwords=custom_stopwords,
    max_words=200,
    max_font_size=100,
    random_state=42,
    width=800,
    height=800,
    color_func=light_green_color_func
)
wc_strengths.generate(strengths_text)

# ------------------ 生成劣势词云（浅红色文字）------------------
wc_weaknesses = WordCloud(
    background_color='white',
    mask=mask_down,
    stopwords=custom_stopwords,
    max_words=200,
    max_font_size=100,
    random_state=42,
    width=800,
    height=800,
    color_func=light_red_color_func     # 固定浅红色
)
wc_weaknesses.generate(weaknesses_text)

# ------------------ 显示两个词云 ------------------
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(wc_strengths, interpolation='bilinear')
plt.axis('off')
plt.title('Strengths', fontsize=16, color='white')  # 标题颜色调亮以适应深色背景

plt.subplot(1, 2, 2)
plt.imshow(wc_weaknesses, interpolation='bilinear')
plt.axis('off')
plt.title('Weaknesses', fontsize=16, color='white')

plt.tight_layout()
plt.show()
```
